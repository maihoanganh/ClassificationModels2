```julia
using CSV
using DataFrames

df = CSV.read("/home/hoanganh/Desktop/math-topics/algebraic_hypersurface_separation/codes/colon_cancer.csv",DataFrame)
```

```julia
df[:,end]
```

```julia
nr,nc=size(df)
D=Matrix{Float64}(undef,nr,nc-1)
for j=1:nr
    for i=2:nc-1
        D[j,i-1]=df[j,i]
    end
    if df[j,nc]=="normal"
        D[j,nc-1]=1
    else
        D[j,nc-1]=2
    end
end
D

N=nc-2
```

```julia
max_col=[maximum(abs.(D[:,j])) for j=1:N]
```

```julia
ind_zero=Vector{Int64}([])
for j=1:N
    if max_col[j]!=0
        D[:,j]/=max_col[j]
    else
        append!(ind_zero,j)
    end
end
ind_zero
```

```julia
D=D[:,setdiff(1:N+1,ind_zero)]
```

```julia
#D[:,1:N].-=0.5
#D[:,1:N]*=2
```

```julia
using LinearAlgebra

max_norm_col=maximum(norm(D[j,1:N]) for j=1:nr) 
```

```julia
R=2
D[:,1:N]/=max_norm_col/R
D
```

```julia
Y=Vector{Matrix{Float64}}(undef,2)

for r in 1:2
    Y[r]=D[findall(u -> u == r, D[:,end]),1:N]
end
```

```julia
Y[2]
```

```julia
t=Vector{Int64}(undef,2)
Y_train=Vector{Matrix{Float64}}(undef,2)

for r=1:2
    t[r]=ceil(Int64,0.6*size(Y[r],1))
    Y_train[r]=Y[r][end-t[r]+1:end,:]
end

Y_train[1]
```

```julia
Y_test=Vector{Matrix{Float64}}(undef,2)

for r=1:2
    Y_test[r]=Y[r][1:end-t[r],:]
end
```

```julia
include("../src/ClassificationModels2.jl")
using .ClassificationModels2


n=N

c=Inf

X=Vector{Matrix{Float64}}(undef,2)
X[1]=Y_train[1]
X[2]=Y_train[2]

k=1

eval_sep_pol=ClassificationModels2.model_SVM(n,X,t,c,k,lamb=0.02,additional_monomials=true)
```

```julia
function classifier(y)
    if eval_sep_pol(y)>0
        return 1
    else
        return 2
    end
end
```

```julia
predict=Vector{Vector{Int64}}(undef,2)

for r=1:2
    predict[r]=[classifier(Y_test[r][j,:]) for j in 1:size(Y_test[r],1)]
end
```

```julia
numcor=Vector{Int64}(undef,2)

for r=1:2
    numcor[r]=length(findall(u -> u == r, predict[r]))
end
```

```julia
sum(numcor)
```

```julia
accuracy=(sum(numcor))/(sum(size(Y_test[r],1) for r=1:2))
```
